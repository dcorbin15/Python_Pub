```python
#7.Which region has the lowest average price for organic avocados?
wide_data = avocado_data.query('type != "organic"').groupby(['region'])['AveragePrice'].aggregate('mean').sort_values(ascending = True)
print(wide_data)
```

# Plots

### Bar Plots

```python
#1.Create a bar plot that shows the mean, median, and standard deviation of the Total Volume column by year 
chart = avocado_data.groupby(['year'])['Total Volume'].agg(['mean', 'median', 'std']).plot.bar()
print(chart)
```

```python
#2.Have the Total Bags sold per year of each type of avocado become more or less consistent over time (3 pts.)?
display(avocado_data.groupby(['type','year'])['Total Bags'] \
        .agg(['std']).plot.bar())
plt.legend(["Organic", "Conventional"])
plt.ylabel('Total Bags')
plt.title('Total Bags Sold over Time')
#Answer: Year over year in the four years that are represented in this dataset shows that for both
# organic and conventional avocados the standard deviation has increased. With standard
# deviation measuring the consistency of data, as that number climbs the consistency decreases.
```

```python
#3. In what year were the most bags of avocados sold ?
most_sold = avocado_data.groupby('year')['Total Bags'].agg('sum').plot.bar()
plt.legend(["Total Bags"])
plt.xlabel('Year')
plt.ylabel('Total Bags')
plt.title('Total Bags Sold over Time')
```

### Scatterplots

```python
#4. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns 
#average_price= data.reset_index
price = sns.scatterplot(data=avocado_data, x= 'AveragePrice', y='Total Volume', hue='type')
price.set(title='Average Price by Volume')
```

```python
#5. Create the same plot but with the hue parameter set to year and the dots for
#   the total volume ranging in size from 10 to 100 
sns.relplot(data=avocado_data, kind='scatter', x='AveragePrice', y='Total Volume', hue='year',
           size='Total Volume', sizes=(10,100))
```

### Cat Plots

```python
#6.Create a plot that looks like the one that follows (2 pts.).
c = sns.catplot(data=avocado_data.query('region in ["TotalUS", "West", "WestTexNewMexico"]'),
                x= 'year', y='Total Volume', kind='bar', hue='region')
```

### Lineplot

```python
#7. Of the three bag sizes, which size sells the most? Hint: Melt the three bag size columns and plot the bags by year (4 pts.).
# Melt the three bag size columns
melted_bagsize = pd.melt(avocado_data, id_vars=['Date', 'year'], value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'], var_name='Bag Size', value_name='Count')

# Plot bagsizes by year
import matplotlib.ticker as ticker
sns.lineplot(x='year', y='Count', hue='Bag Size', data=melted_bagsize) # Used a Seaborn line plot
years = ticker.MultipleLocator(base=1.0)  # Show ticks every 1 year
fmt = ticker.StrMethodFormatter("{x:.0f}")  # Format the tick labels as integers so only whole years are showm
plt.gca().xaxis.set_major_locator(years) # Get current axes [year] 
plt.gca().xaxis.set_major_formatter(fmt) # Get current axes [format] 

plt.show()
```

## DataFrame Manipulation

```python
#1. Convert the Date column to the datetime type (2 pts.).
avocado_data['Date'] = pd.to_datetime(avocado_data['Date'], format='%Y-%m-%d')
```

```python
#2.Notice that the column names are inconsistent. Change the column names so they all use title case with no spaces (6 pts.).
avocado_data.rename(columns={'Unnamed:0': 'Unnamed:0','Date': 'Date','AveragePrice': 'AveragePrice', 
                             'Total Vlume': 'TotalVolume', 'Total Bags': 'TotalBags', 'Small Bags': 'SmallBags',
                             'Large Bags': 'LargeBags', 'XLarge Bags': 'XLargeBags','type': 'Type', 'year': 'Year',
                             'region': 'Region'}, inplace=True)
avocado_data.head()
```

```python
#3.Add a column with the percentage of the bags in the TotalBags column that are extra-large bags 

#Step 1: Find the sum of XLargeBags and TotalBags columns 
#avocado_data['XLargeBags'].sum()
#avocado_data['TotalBags'].sum() 

#Step 2: [ADD Column] & Calculate percentage of XLarge Bags in the Total Bags column 
avocado_data["XLBag_Pct"] = avocado_data['XLargeBags'].sum() / avocado_data['TotalBags'].sum() * 100
avocado_data
```

```python
#4.Add a column with the percentage of the bags in the TotalBags column that are large bags (2 pts.).
#Step 1: Find the sum of XLargeBags and TotalBags columns 
#avocado_data['LargeBags'].sum()
#avocado_data['TotalBags'].sum() 

#Step 2: [ADD Column] & Calculate percentage of Large Bags in the Total Bags column 
avocado_data["LBag_Pct"] = avocado_data['LargeBags'].sum() / avocado_data['TotalBags'].sum() * 100
avocado_data
```

```python
#5.Add a column with the percentage of the bags in the TotalBags column that are small bags (2 pts.).
#Step 2: Find the sum of XLargeBags and TotalBags columns 
#avocado_data['SmallBags'].sum()
#avocado_data['TotalBags'].sum() 

#Step 3: [ADD Column] & Calculate percentage of Small Bags in the Total Bags column 
avocado_data["SmallBag_Pct"] = avocado_data['SmallBags'].sum() / avocado_data['TotalBags'].sum() * 100
avocado_data
```

```python
#6.Display the first five rows of data to view the new columns (2 pts.).
avocado_data.head()
```

```python
#7.Assign the Region, Type, Year, and TotalBags columns to a new DataFrame (2 pts.).
new_df = pd.DataFrame(avocado_data[['Region','Type', 'Year', 'TotalBags']])
new_df
```

```python
#8.Group, sum, and unstack this data to produce a new DataFrame that looks like this (3 pts.):
df_grouped = avocado_data.groupby(['Region','Type', 'Year']).sum()[['TotalBags']].unstack(level=1)
df_grouped
```

### Analyze Data in new View

```python
#1.Display the unique values in the region column, and notice that one of the regions is TotalUS 
#avocado_data.region.unique() --> Simple Array View
unique_region = avocado_data['Region'].unique() #--> Detailed Numbered View
s = pd.Series(unique_region)
s_with_index = s.reset_index(drop=True)
print(s_with_index)
```

```python
#2. Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column.
#   Name the column that contains the type of bag Bags, and name the column that contains the number of bags Count.
#   Then, display the resulting new DataFrame (4 pts.).

# Melt the data in the Bags and Small Bags columns
small_tot_melted = pd.melt(avocado_data, id_vars=['Date'],
                    value_vars=['TotalBags', 'SmallBags'],
                    var_name='Bags', value_name='Count')

# Display the resulting DataFrame
print(small_tot_melted.head())
```

```python
#3. Plot the melted data with Seaborn in a line plot, using the hue parameter to distinguish between the bag types (2 pts.).
#I imported matplotlib to rotate the date variables because they were overlapping p.131 
sns.lineplot(x='Date', y='Count', hue='Bags', data=small_tot_melted)
plt.xticks(rotation=30)
plt.show()
```
