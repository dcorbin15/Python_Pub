```python
#5. Index the data on the Date column. Then, generate time periods for the first day of each year from 1950 to 2020 (3 pts.).
#Index
gold_data.set_index('Date', inplace=True)
# Generate time periods for the first day of each year from 1950 to 2020
start_date = pd.Timestamp('1950-01-01')
end_date = pd.Timestamp('2020-01-01')
year_starts = pd.date_range(start=start_date, end=end_date, freq='YS')

# Print the first 5 values of the year_starts variable
print(year_starts[:5])
```

```python
#6. Reindex the data using the new time periods,
#   and display the first five rows again to see that there is just one row for each year (2 pts.).
gold_data = gold_data.reindex(year_starts)
gold_data.head()
```

```python
#7. Plot the reindexed data with a Pandas line plot (2 pts.).
gold_data.reindex(year_starts).plot()
```

```python
#8. Resample the data to a quarterly frequency and plot the data again (2 pts.).
#Resample to quarterly frequency
gold_data_quarterly = gold_data.resample(rule='Q').sum()

# Plot the data
gold_data_quarterly.plot()
plt.title('Gold Prices Quarterly')
plt.show()
```

```python
#9. Resample the data to a yearly frequency and plot the data one more time (2 pts.).
#Resample to yearly frequency
gold_data_yearly = gold_data.resample(rule='Y').sum()

# Plot the data
gold_data_yearly.plot()
plt.title('Gold Prices Yearly')
plt.show()
```

```python
#10. Plot the rolling mean for the data using a Pandas line plot (2 pts.).
#    Experiment with two different values for the window parameter,
#    include each different window value plot in a separate cell (4 pts.).

# Example 1:

# Calculate the rolling mean with a window of 12 months
rolling_mean = gold_data.rolling(window=12).mean()

# Plot the rolling mean
rolling_mean.plot()
plt.title('Gold Prices Rolling Mean')
plt.show()
```

```python
# Part 1 Example 2:
# Calculate the rolling mean with a window of 3 months
rolling_mean2 = gold_data.rolling(window=3).mean()

# Plot the rolling mean
rolling_mean2.plot()
plt.title('Gold Prices Rolling Mean 2')
plt.show()
```

```python
# Part 2: Include each different window value plot in a separate cell (4 pts.).
def plot_rolling_mean(window_size):
    # Calculate the rolling mean with the specified window size
    rolling_mean = gold_data.rolling(window=window_size).mean()
    
    # Plot the rolling mean using Seaborn
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=rolling_mean, x=rolling_mean.index, y='Price')
    plt.title(f'Rolling Mean ({window_size}-Month Window)')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()
```

```python
# Plot the rolling mean with a 3-month window
plot_rolling_mean(3)
# Plot the rolling mean with a 6-month window
plot_rolling_mean(6)
# Plot the rolling mean with a 12-month window
plot_rolling_mean(12)
```
