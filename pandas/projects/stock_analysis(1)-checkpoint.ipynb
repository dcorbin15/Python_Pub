# Import Data

```python
data = yf.download(
        tickers = ['AAPL', 'MSFT','GOOG', 'META'],
        period = '52wk',
        interval = '1d'
    )
prices = data['Adj Close']
print(f"Retrieved {len(prices)} data points from Yahoo Finance")
```

# Create New DataFrame

```python
new_data = []
tickers = ['AAPL', 'MSFT', 'GOOG','META']
for ticker in tickers:
 tkr = yf.Ticker(ticker)
 hist = tkr.history(period='5d').reset_index()
 records = hist[['Date','Close']].to_records(index=False)
 records = list(records)
 d = records
 records = [(ticker, str(elem[0])[:10], elem[1]) for elem in records]
 new_data = new_data + records 
print(new_data)
```
