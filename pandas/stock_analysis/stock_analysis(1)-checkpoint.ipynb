# Processing Multvariate Time Series

```python
stocks= pd.DataFrame()
tickers= ['MSFT', 'TSLA', 'AAPL', 'ORCL', 'AMZN']
```

```python
for ticker in tickers:
    tkr = yf.Ticker(ticker)
    hist= tkr.history(period='5d')
    hist= pd.DataFrame(hist[['Close']].rename(columns={'Close': ticker}))
    if stocks.empty:
        stocks = hist
    else:
        stocks = stocks.join(hist)
```

```python
# Resulting DataFrame
stocks
```

```python
stocks_to_keep = []
for i in stocks.columns:
    if stocks[stocks[i].shift(1)< .97].empty:
        stocks_to_keep.append(i)
print(stocks_to_keep)
```

```python
print(stocks[stocks_to_keep])
```

# Analyzing Dependencies between variables

```python
import numpy as np 
ticker = 'TSLA'
tkr = yf.Ticker(ticker)
df = tkr.history(period='1mo')
```

```python
df = df[['Close', 'Volume',]].rename(columns={'Close': 'Price'})
```

```python
df['priceRise'] = np.log(df['Price'] / df['Price'].shift(1))
```

```python
df['volumeRise'] = np.log(df['Volume'] / df['Volume'].shift(1))
```

```python
df
```

```python
print(df[abs(df['priceRise']) > .05])
```

```python
mean_volumeRise = df['volumeRise'].mean()
if isinstance(mean_volumeRise, float):
    print(np.round(mean_volumeRise, 4))
```

```python
print(round(df[abs(df['priceRise']) > .05] ['volumeRise'].mean(), 4))
```
